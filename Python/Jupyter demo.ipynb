```python
from scipy import misc #Import misc module from scipy (general science library)
import numpy as np #Import numpy (matrix/array operations library)
%matplotlib inline 
#Jupyter magic command to show images inline
import matplotlib.pyplot as plt #Import matplotlib plotting library to display arrays
from IPython.display import Image #Import Image utility to display image from file
import skimage
import scipy
```

```python
#Save filename to variable f
f = 'dog.jpg'
```

```python
#Display image
Image(f)
```

```python
#Read image as an array
im = scipy.misc.imread(f)
im = skimage.img_as_float(im)
```

```python
#See the array's shape
im.shape
```

```python
#Inspect the array
im
```

```python
#Look at just the green channel
im[:,:,1]
```

```python
#Mean value of the image
np.mean(im)
```

```python
#Show the image
plt.imshow(im)
```

```python
#Cut out most of the green channel
im[:,:,0] = im[:,:,0] * 0.05
```

```python
#Show the image again
plt.imshow(im)
```
